```python
print("The luminosity distance at redshift 1 predicted by iCosmos is 6904.2185 Mpc.")
print("The luminosity distance at redshift 1 predicted by us is", bg.luminosity_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa), "Mpc.")
print("Percentual difference is", 100*(1-bg.luminosity_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/6904.2185),".")
```

# Checks: open Universe
For this checks I assume the following values of the cosmological parameters:
- $\Omega_M=0.32$
- $\Omega_R=0.0$
- $\Omega_{DE}=0.48$
- $\Omega_{K}=0.2$
- $H_0=67.0\,\textrm{km}\,\textrm{s}^{-1}\,\textrm{Mpc}^{-1}$
- $w_0=-1.1$
- $w_a = 0.2$

```python
c0 = 2.99792458e+5
H0 = 67
ΩM = 0.32
ΩDE = 0.48
ΩK = 1-ΩM-ΩDE
w0 = -1.1
wa = 0.2
```

```python
print("The comoving distance at redshift 1 predicted by iCosmos is 3255.9978 Mpc.")
print("The comoving distance at redshift 1 predicted by us is", bg.comoving_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa), "Mpc.")
print("Percentual difference is", 100*(1-bg.comoving_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/3255.9978),".")
```

```python
print("The angular diameter distance at redshift 1 predicted by iCosmos is 1656.8863 Mpc.")
print("The angular diameter distance at redshift 1 predicted by us is", bg.angular_diameter_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa) , "Mpc.")
print("Percentual difference is", 100*(1-bg.angular_diameter_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/1656.8863),".")
```

```python
print("The luminosity distance at redshift 1 predicted by iCosmos is 6627.5453 Mpc.")
print("The luminosity distance at redshift 1 predicted by us is", bg.luminosity_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa), "Mpc.")
print("Percentual difference is", 100*(1-bg.luminosity_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/6627.5453),".")
```

```python
x = c0/H0/np.sqrt(ΩK)
z = 1
new_angular_diameter = x*np.sinh(bg.comoving_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/x)/(1+z)
print("The angular diameter distance at redshift 1 predicted by new algorithm is", new_angular_diameter, "Mpc.")
print("The percentual difference between the new angular diameter distance and iCosmos is", 100*(1-new_angular_diameter/1656.8863),".")
```

```python
x = c0/H0/np.sqrt(ΩK)
z = 1
new_luminosity = x*np.sinh(bg.comoving_distance(z=1., H0 = H0, ΩM=ΩM, ΩDE=ΩDE, w0= w0, wa = wa)/x)*(1+z)
print("The luminosity distance at redshift 1 predicted by new algorithm is", new_luminosity, "Mpc.")
print("The percentual difference between the new luminosity distance and iCosmos is", 100*(1-new_luminosity/6627.5453),".")
```
